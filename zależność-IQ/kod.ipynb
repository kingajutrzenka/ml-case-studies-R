```r
plot(avg_income~IQ, data = iq_data, 
        main = "Zależność średniego przychodu w kraju od IQ", col="#D13654")
```

```r
cor.test(iq_data$IQ, iq_data$avg_income)
```

korelacja między poziomem iq, a średnim przychodem w kraju wynosi około 0.57, więc jest to silna korelacja dodatnia

```r
plot(avg_temp~IQ, data = iq_data, 
        main = "Zależność średniej temperatury w kraju od IQ", col="#D13654")
```

```r
cor.test(iq_data$IQ, iq_data$avg_temp)
```

korelacja między poziomem iq, a średnią temperaturą w kraju wynosi około -0.63, więc jest to silna korelacja ujemna

```r
plot(education_expenditure~IQ, data = iq_data, 
        main = "Zależność kwoty edukacji w kraju od IQ", col="#D13654")
```

```r
cor.test(iq_data$education_expenditure, iq_data$IQ)
```

korelacja między poziomem iq, a kosztem edukacji w kraju wynosi około 0.57, więc jest to silna korelacja dodatnia
